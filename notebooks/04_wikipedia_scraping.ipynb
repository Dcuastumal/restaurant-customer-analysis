# Chicago Cultural Analysis

## Objective

The objective of this notebook is to collect external cultural information about the selected city using web scraping techniques.

Unlike the Yelp Fusion API, Wikipedia provides descriptive information that enriches the business analysis by adding historical, cultural, and tourism context.

The extracted information will complement the customer and restaurant analyses performed in the previous stages of the project.

```python
# Import Libraries

import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# ==========================================
# Wikipedia Scraping
# ==========================================

url = "https://en.wikipedia.org/wiki/Chicago"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

print(response.status_code)

soup = BeautifulSoup(response.text, "html.parser")

paragraphs = soup.find_all("p")

culture_text = []

for paragraph in paragraphs[:10]:

    text = paragraph.get_text(strip=True)

    if len(text) > 80:
        culture_text.append(text)

chicago_culture = pd.DataFrame({
    "culture_information": culture_text
})

chicago_culture.head()
```

## Results

The scraping process successfully extracted descriptive information from the official Wikipedia page for Chicago.

The extracted content includes information about the city's history, economy, culture, tourism, and urban development.

Although these data are unstructured, they provide valuable context that complements the customer and restaurant analyses developed throughout the project.

```python
# ==========================================
# Export Results
# ==========================================

chicago_culture.to_csv(
    "../data/chicago_culture.csv",
    index=False
)
```

# Conclusions

The scraping process successfully extracted cultural information from Wikipedia about the selected city.

Unlike structured customer and restaurant datasets, Wikipedia provides descriptive information that enriches the business analysis by adding historical, cultural, and tourism context.
